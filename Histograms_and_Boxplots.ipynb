```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
filename = '/content/drive/MyDrive/catsvdogs.xlsx'
df = pd.read_excel('/content/drive/MyDrive/catsvdogs.xlsx')
df
```

```python
#1. Plot a histogram of the Percentage of Dog Owners.
plt.hist(df['Percentage of Dog Owners']);        
```

```python
#2. Plot a histogram of the Percentage of Cat Owners.
plt.hist(df['Percentage of Cat Owners']);
```

```python
#.3a. What can we see by comparing these two histograms? What information does this tell us?
#ans: By two these histograms i observed that the percentage of dog owners is more than that of the percentage of cat owners. And highest value is in percentage of ccat owners.
```

```python
#3b. Optional/Bonus: plot these two histograms on the same plot. An example of what this could look like is given below:
plt.hist(df['Percentage of Cat Owners'])
plt.hist(df['Percentage of Dog Owners']);
```

```python
#4. Create two boxplots on the same plot: one of the mean number of cats per household and the other of the mean number of dogs.
dogs = df['Mean Number of Dogs per household']
cats = df['Mean Number of Cats']
plt.boxplot([cats,dogs], notch=True, 
            labels=['Dogs', 'Cats'],
            widths = .7,
            patch_artist=True,
            medianprops = dict(linestyle='-', linewidth=2, color='Yellow'),
            boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'blue', alpha = .4));
plt.ylabel('Mean Number of Animals per state', fontsize = 20);
```

```python
# 5. What can we see by comparing these two boxplots? What information does this tell us?
# In this two box plots i observed that the dogs mean number of animals per state is more than that of the cat mean number. And at the same time median value of dogs is more than that of the cats median. 
```
